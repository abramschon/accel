# Check encodings
This notebook is to go through all the variables and see whether our encodings of them are sensible

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from accel.read_data import load_data, get_numeric_cols, get_object_cols, custom_encodings
from accel.constants import DATA_DIR
```

```python
raw_data = load_data(DATA_DIR / "shared/reduced-cohort.csv") # can also see biobank-encoded-cohort
```

```python
raw_data.shape
```

```python
cat_cols = get_object_cols(raw_data)
num_cols = get_numeric_cols(raw_data)
```

```python
raw_data[num_cols].head()
```

```python
raw_data[cat_cols].head()
```

```python
data_enc = custom_encodings(raw_data)
```

```python
cat_cols = get_object_cols(data_enc)
num_cols = get_numeric_cols(data_enc)
```

```python
data_enc[cat_cols].head()
```

```python
data_enc[num_cols].head()
```

```python
for col in cat_cols:
    print(col)
    print(data_enc[col].unique())
    data_enc[col].value_counts().plot(kind='bar')
    plt.show()
```

```python
for col in num_cols:
    print(col)
    data_enc[col].plot(kind='hist', bins=100)
    plt.show()
```
